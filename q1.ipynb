```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Image Creation

height, width = 200, 200
image = np.full((height, width), 255, dtype=np.uint8) 
cv2.rectangle(image, (50, 50), (75, 150), 0, -1)
cv2.rectangle(image, (125,75 ), (175, 125), 125, -1)

# Add Gaussian noise
mean = 0
stddev = 30
gaussian_noise = np.random.normal(mean, stddev, image.shape)
noisy_image = image.astype(np.float32) + gaussian_noise
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

# Apply Otsu's thresholding
ret, otsu_thresh = cv2.threshold(noisy_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Plot results
plt.axis(False)
plt.imshow(image, cmap='gray')
plt.show()

plt.axis(False)
plt.imshow(noisy_image, cmap='gray')
plt.show()

plt.axis(False)
plt.imshow(otsu_thresh, cmap='gray')
plt.show()
```
